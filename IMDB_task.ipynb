### Number of nan  values in each column

```python
movies.isna().sum()
```

```python
# fill missing values
movies.Certificate.interpolate(method="pad",inplace=True)
movies.Gross.fillna(0,inplace=True)
movies.Meta_score.interpolate(inplace=True)
```

```python
#make sure of the above step is correct
movies.isna().sum()
```

### Genre Counts unique and normal

```python
#get count for each unique Genre
genre_unique_counts = movies.Genre.map(lambda genre: genre.split(", ")).explode().value_counts()
```

```python
#get count for each Genre as it is. e.g single Genre: "Action, Drama, Romance"
genre_counts  = movies.groupby("Genre").Genre.value_counts().sort_values()
```

### Genre counts charts

```python
# plotting Genres as bar chart
genre_counts.plot(x='genre', y='count', kind='bar', figsize=(30, 20))
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre of Movies and TV Shows')
plt.show()
```

```python
#plotting unique Genres as bar chart
genre_unique_counts.plot(x='genre', y='count', kind='bar', figsize=(10, 8))
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre of Movies and TV Shows')
plt.show()
```

```python
movies.Director.unique().size
```

### Movies by Director

```python
movies_count_by_director =movies.groupby("Director").Director.value_counts().sort_values()
```

```python
#plotting unique Genres as bar chart
movies_count_by_director.plot(x='director', y='movies/TV shows count', kind='bar', figsize=(40, 20))
plt.xlabel('director')
plt.ylabel('movies/TV shows count')
plt.title('Most Popular Genre of Movies and TV Shows')
plt.show()
```

### Valuable values

```python
#remove comma from gross string and convert it to float
movies.Gross = movies.Gross.map(lambda gross: str(gross).replace(",","")).astype("float")
# get the movie with the highest Gross and select the title, Genre, Gross and director
movies.Gross.idxmax(), movies.loc[movies.Gross.idxmax(),["Series_Title","Genre","Gross","Director"]]
```

```python
movies.drop(movies[movies.Released_Year=="PG"].index, inplace=True)
movies.Released_Year.astype("int64").agg([min, max])
```

```python
(topGenre,topGenreCount) = genre_counts.idxmax(),genre_counts[genre_counts.idxmax()]
(topGenre,topGenreCount)
```

```python
(topGenre,topGenreCount, bottomGenre, bottomGenreCount)=genre_unique_counts.idxmax(),genre_unique_counts[genre_unique_counts.idxmax()],genre_unique_counts.idxmin(),genre_unique_counts[genre_unique_counts.idxmin()]
(topGenre,topGenreCount, bottomGenre, bottomGenreCount)
```

```python
(director, maxMoviesCountPerDirector) = movies_count_by_director.idxmax(), movies_count_by_director[movies_count_by_director.idxmax()]
(director, maxMoviesCountPerDirector)
```
